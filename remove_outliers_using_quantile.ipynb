**Removing outlier using Quantile**

```python
percentile_99 = df.Income.quantile(0.99)
percentile_99
```

```python
df_outlier = df[df.Income > percentile_99]
df_outlier
```

```python
df_no_outlier = df[df.Income <= percentile_99]
df_no_outlier
```
